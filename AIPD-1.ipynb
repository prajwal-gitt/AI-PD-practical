```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv("diabetes.csv")
```

```python
df
```

```python
df.info()
```

```python
df. describe()
```

```python
df = df.drop(columns=['Pregnancies','SkinThickness','Insulin','DiabetesPedigreeFunction'],axis = 1)
```

```python
df
```

```python
df.isnull().sum()
```

```python
sns.pairplot(df)
```

```python
x = df.drop(columns=['Outcome'])
```

```python
y = df['Outcome']
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x,y,train_size = 0.8 , random_state = 42)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score , classification_report , confusion_matrix , ConfusionMatrixDisplay
```

```python
accuracy_LR = accuracy_score(y_test , y_pred)
cr = classification_report(y_test , y_pred)
cm = confusion_matrix(y_test , y_pred)
```

```python
accuracy_LR
```

```python
cm
```

```python
print(cr)
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix = cm)
disp.plot()
plt.show()
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model_GNB = GaussianNB()
```

```python
model_GNB
```

```python
model_GNB.fit(x_train,y_train)
```

```python
y_pred_GNB = model_GNB.predict(x_test)
```

```python
y_pred_GNB
```

```python
accuracy_GNB = accuracy_score(y_test , y_pred_GNB)
cr = classification_report(y_test , y_pred_GNB)
cm = confusion_matrix(y_test , y_pred_GNB)
```

```python
accuracy_GNB
```

```python
print(cr)
```

```python
cm
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix = cm)
disp.plot()
plt.show()
```

```python
from sklearn.svm import SVC
```

```python
model_SVM = SVC()
```

```python
model_SVM
```

```python
model_SVM.fit(x_train,y_train)
```

```python
y_pred_SVM = model_SVM.predict(x_test)
```

```python
y_pred_SVM
```

```python
accuracy_SVM = accuracy_score(y_test , y_pred_SVM)
cr = classification_report(y_test , y_pred_SVM)
cm = confusion_matrix(y_test , y_pred_SVM)
```

```python
accuracy_SVM
```

```python
print(cr)
```

```python
cm
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix = cm)
disp.plot()
plt.show()
```

```python
import matplotlib.pyplot as plt

models = ['Linear Regression', 'Gaussian NB', 'SVM']
accuracies = [accuracy_LR, accuracy_GNB, accuracy_SVM]

colors = ['blue', 'green', 'orange']

plt.bar(models, accuracies, color=colors)
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.title('Model Accuracy Comparison')

plt.show()
```
